# Parsing messy CA precincts

Creates 3 CSV files: unique precincts, absentee precincts, and "SOV" data. To simply find the number of precincts in each county, the file we are interested in is the unique precincts one. 

```python
import geopandas as gp
import pandas as pd
import os

#precincts = gp.read_file('/Users/kiwan/Downloads/.shp')
from_site = pd.read_csv('./2018_CA/from_site/state_p18_sov_data_by_p18_svprec.csv')
from_shp = pd.read_csv('./2018_CA/from_shapefile.csv')
df = pd.DataFrame(from_site)

pct = []
pct_abs = []
pct_sov = []
```

```python
df["SVPREC"] = df["SVPREC"].astype(str) 

for i in range(len(df["SVPREC"])):
    uuid = df["SVPREC"][i]
    if (uuid.endswith('A')):
        pct_abs.append(df.loc[i])
    elif (uuid.startswith('SOV')):
        pct_sov.append(df.loc[i])
    else:
        pct.append(df.loc[i])
        
pct_df = pd.DataFrame(pct)
pct_abs_df = pd.DataFrame(pct_abs)
pct_sov_df = pd.DataFrame(pct_sov)

```

```python
# Saved to the same directory that this notebook was launched from

pct_df.to_csv('2018_CA_pct.csv')
pct_abs_df.to_csv('2018_CA_absentee_pct.csv')
pct_sov_df.to_csv('2018_CA_SOV_pct.csv')
```

```python
import numpy as np

num_precincts = pct_df.groupby("FIPS")
table = num_precincts.agg(np.size)
print(table)
```
